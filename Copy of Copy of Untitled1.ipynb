```python
import plotly.graph_objects as go
```

```python
df = pd.read_csv('TATAPOWER.NS (3).csv')
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```

```python
result = talib.CDLTRISTAR(df['Open'],df['High'] ,df['Low'],df['Close'])
```

```python
plt.title('TRISTAR PATTERN')
plt.xlabel('Date')
plt.ylabel('TRISTAR PATTERN')
plt.plot(df['Date'],result,lw=1, label='INDEX')
plt.legend()

plt.show()
```
